# **Clustering**

 What is unsupervised learning in the context of machine learning
 - Unsupervised learning is a type of machine learning in which a model is trained on data that has no labeled outputs, meaning the algorithm must discover patterns and structure on its own. Instead of predicting a known target, it analyzes similarities, differences, and relationships within the data to group, organize, or reduce it. Common tasks in unsupervised learning include clustering (such as grouping customers with similar behavior), association rule mining, and dimensionality reduction (like identifying the most important features). This approach is especially useful when dealing with large, unstructured datasets where prior labeling is difficult or unavailable.

 How does K-Means clustering algorithm work
 - K-Means is an unsupervised learning algorithm used to group data into K clusters based on similarity. First, the value of K (number of clusters) is chosen, and K centroids are initialized, usually at random. Each data point is then assigned to the nearest centroid using a distance measure such as Euclidean distance. After all points are assigned, the centroids are recalculated as the mean of all points in each cluster. This process of assignment and centroid update is repeated iteratively until the centroids no longer change significantly or the assignments stabilize. The final result is K clusters where data points within the same cluster are more similar to each other than to those in other clusters.

Explain the concept of a dendrogram in hierarchical clustering
- A dendrogram is a tree-like diagram used in hierarchical clustering to visually represent how data points are grouped together at different levels of similarity or distance. Each leaf of the dendrogram represents an individual data point, and branches show how clusters are merged (in agglomerative clustering) or split (in divisive clustering) as the algorithm progresses. The height of the branches indicates the distance or dissimilarity at which clusters are joined. By drawing a horizontal cut across the dendrogram at a chosen height, we can decide the number of clusters and see which data points belong to each cluster.

What is the main difference between K-Means and Hierarchical Clustering
- The main difference between K-Means and Hierarchical Clustering lies in how clusters are formed and specified. K-Means requires the number of clusters (K) to be defined in advance and iteratively assigns data points to the nearest centroid to optimize cluster compactness. In contrast, Hierarchical Clustering does not require a predefined number of clusters; instead, it builds a hierarchy of clusters represented by a dendrogram, from which the number of clusters can be chosen later. Additionally, K-Means is generally faster and more suitable for large datasets, while hierarchical clustering is more computationally expensive but provides a more detailed view of data relationships.

What are the advantages of DBSCAN over K-Means
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means, especially for complex and real-world datasets. Unlike K-Means, DBSCAN does not require the number of clusters to be specified in advance and can automatically discover clusters based on data density. It is able to find arbitrarily shaped clusters, whereas K-Means works best with spherical clusters. DBSCAN also identifies noise and outliers explicitly, labeling them as points that do not belong to any cluster, while K-Means forces every point into a cluster. Additionally, DBSCAN is more robust to outliers and performs well when clusters have clear density differences, making it suitable for spatial and geospatial data.

When would you use Silhouette Score in clustering
- You would use the **Silhouette Score** in clustering to **evaluate the quality of clusters** and to help **choose the optimal number of clusters**. It measures how similar a data point is to its own cluster compared to other clusters, combining both **cluster cohesion** (how close points are within a cluster) and **cluster separation** (how far apart clusters are). The score ranges from **–1 to +1**, where a higher value indicates better-defined clusters. Silhouette Score is commonly used with algorithms like **K-Means** to compare different values of *K* and select the one that produces the most meaningful clustering structure.

What are the limitations of Hierarchical Clustering
- Hierarchical clustering has several limitations despite its interpretability. It is **computationally expensive** in both time and memory, making it unsuitable for very large datasets. Once a merge or split is made, it **cannot be undone**, so early incorrect decisions may affect the final clustering result. The method is also **sensitive to noise and outliers**, which can distort the cluster hierarchy. Additionally, the quality of results depends heavily on the **choice of distance metric and linkage method** (such as single, complete, or average linkage), and there is **no clear objective function** to optimize cluster quality.

Why is feature scaling important in clustering algorithms like K-Means
- Feature scaling is important in clustering algorithms like **K-Means** because these algorithms rely on **distance calculations** (typically Euclidean distance) to group data points. If features are on different scales, variables with larger ranges can **dominate the distance measure**, causing the clustering to be biased toward those features and leading to incorrect or misleading clusters. Scaling ensures that all features contribute **equally** to the distance computation, improving cluster accuracy and stability. Techniques such as **standardization** or **min–max normalization** are therefore commonly applied before performing K-Means clustering.

How does DBSCAN identify noise points
- DBSCAN identifies noise points based on **data density** using two key parameters: **ε (epsilon)** and **MinPts**. A point is considered a **core point** if it has at least *MinPts* data points (including itself) within its ε-neighborhood. A **border point** lies within the ε-neighborhood of a core point but has fewer than *MinPts* neighbors itself. Any point that is **neither a core point nor a border point**—meaning it does not fall within the ε-neighborhood of any core point—is labeled as a **noise (outlier) point**. This allows DBSCAN to explicitly separate sparse, isolated points from dense clusters.

Define inertia in the context of K-Means
- In the context of **K-Means**, **inertia** is a measure of how well the data points are clustered around their respective centroids. It is defined as the **sum of squared distances** between each data point and the centroid of the cluster to which it is assigned. Lower inertia indicates that data points are closer to their centroids, meaning the clusters are more compact. Inertia is commonly used in the **elbow method** to help determine an appropriate number of clusters, although it always decreases as the number of clusters increases.

What is the elbow method in K-Means clustering
- The **elbow method** is a technique used in **K-Means clustering** to determine the optimal number of clusters (**K**). It involves running the K-Means algorithm for a range of K values and plotting **inertia** (the sum of squared distances between data points and their cluster centroids) against K. As K increases, inertia decreases, but at a certain point the rate of decrease slows down significantly, forming an “elbow” in the plot. This elbow point represents a balance between model simplicity and cluster compactness and is often chosen as the optimal number of clusters.

Describe the concept of "density" in DBSCAN
- In **DBSCAN**, **density** refers to how closely data points are packed together in a given region of the data space. It is defined using two parameters: **ε (epsilon)**, which specifies the radius of a neighborhood around a point, and **MinPts**, the minimum number of points required within that ε-neighborhood. A region is considered **dense** if a point has at least *MinPts* neighbors within distance ε, making it a **core point**. DBSCAN forms clusters by connecting these dense regions, while areas with insufficient density are treated as **noise or outliers**.

Can hierarchical clustering be used on categorical data
- Yes, **hierarchical clustering can be used on categorical data**, but not directly with standard distance measures like Euclidean distance. Since categorical variables do not have a natural numeric order, you must use **appropriate similarity or distance measures** such as **Hamming distance**, **Jaccard similarity**, or **Gower distance**, which can handle categorical or mixed data types. Additionally, encoding techniques like **one-hot encoding** may be applied, though they can increase dimensionality. With the right distance metric and linkage method, hierarchical clustering can effectively group categorical data.

What does a negative Silhouette Score indicate
- A **negative Silhouette Score** indicates that a data point is likely **misclassified**—it is closer to points in a neighboring cluster than to points in its own cluster. This suggests **poor cluster separation**, meaning the clusters overlap significantly or are not well defined. When many points have negative silhouette values, it often implies that the chosen number of clusters is inappropriate or that the clustering algorithm is not well suited to the structure of the data.

Explain the term "linkage criteria" in hierarchical clustering
- In **hierarchical clustering**, **linkage criteria** define how the distance between two clusters is calculated when deciding which clusters to merge or split. It determines the rule for measuring similarity between clusters based on the distances between their data points. Common linkage methods include **single linkage** (minimum distance between points of two clusters), **complete linkage** (maximum distance), **average linkage** (average distance), and **Ward’s linkage** (minimizes the increase in total within-cluster variance). The choice of linkage criteria significantly affects the shape, compactness, and interpretability of the resulting cluster hierarchy.

Why might K-Means clustering perform poorly on data with varying cluster sizes or densities
- K-Means clustering may perform poorly on data with varying cluster sizes or densities because it assumes that clusters are spherical and of roughly equal size. The algorithm assigns points to the nearest centroid based on distance, so in cases where clusters have different densities, K-Means can:

- Split large or dense clusters into multiple smaller clusters.

- Merge small or sparse clusters into nearby larger clusters.

- Misclassify points near the boundaries of clusters with different densities.

- Because K-Means relies solely on distance to centroids, it struggles with clusters that are irregularly shaped, unevenly sized, or have non-uniform point distributions, unlike density-based algorithms like DBSCAN, which can handle such scenarios more effectively.

 What are the core parameters in DBSCAN, and how do they influence clustering
 - The core parameters in **DBSCAN** are **ε (epsilon)** and **MinPts**, and they play a crucial role in defining clusters:

1. **ε (epsilon)**: This defines the **radius of the neighborhood** around each data point. Points within this distance are considered neighbors. A small ε may result in many points being labeled as noise, while a large ε can merge distinct clusters into one.

2. **MinPts**: This is the **minimum number of points required** within a point’s ε-neighborhood for it to be considered a **core point**. A higher MinPts value makes clustering stricter, potentially leaving more points as noise, while a lower value allows smaller clusters to form.

- Together, these parameters control **cluster density, shape, and noise detection**, directly influencing the size, number, and quality of clusters DBSCAN produces.

How does K-Means++ improve upon standard K-Means initialization
- K-Means++ improves standard K-Means by providing a smarter initialization of cluster centroids to reduce the chances of poor clustering and slow convergence. In standard K-Means, centroids are chosen randomly, which can lead to suboptimal clusters or longer iterations. K-Means++ selects the first centroid randomly but chooses subsequent centroids probabilistically based on their distance from existing centroids—points farther from existing centroids are more likely to be chosen. This spreads the initial centroids more evenly across the data, resulting in faster convergence and often better final clustering compared to random initialization.

What is agglomerative clustering
- **Agglomerative clustering** is a type of **hierarchical clustering** that builds a cluster hierarchy in a **bottom-up approach**. Initially, each data point is treated as its **own cluster**. Then, at each step, the **two closest clusters** are merged based on a chosen **distance metric** and **linkage criterion** (such as single, complete, or average linkage). This process continues iteratively until **all points are merged into a single cluster** or a stopping condition is reached. The results are often visualized using a **dendrogram**, which shows the sequence of merges and helps determine the optimal number of clusters.

What makes Silhouette Score a better metric than just inertia for model evaluation
- The **Silhouette Score** is often considered better than just **inertia** for evaluating clustering because it accounts for both **cohesion** and **separation**, whereas inertia only measures **within-cluster compactness**. Specifically:

1. **Cohesion**: Silhouette Score considers how close each point is to other points in its own cluster.
2. **Separation**: It also accounts for how far each point is from points in the nearest neighboring cluster.

- Inertia, on the other hand, always **decreases as the number of clusters increases**, making it hard to judge the true quality of clustering. Silhouette Score, ranging from –1 to +1, provides a **normalized measure** that reflects both how well points fit their own clusters and how distinct clusters are from each other, giving a more meaningful assessment of cluster quality.


 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=1.0, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Visualize the clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering on Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Generate synthetic moon-shaped data
X, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)

# Step 2: Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_db = dbscan.fit_predict(X)

# Step 3: Identify core points, clusters, and outliers
core_samples_mask = np.zeros_like(y_db, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
labels = y_db
outliers = labels == -1  # DBSCAN labels noise points as -1

# Step 4: Visualize clusters and outliers
plt.figure(figsize=(8,6))
plt.scatter(X[labels != -1, 0], X[labels != -1, 1], c=labels[labels != -1], cmap='viridis', s=50, label='Cluster points')
plt.scatter(X[outliers, 0], X[outliers, 1], c='red', s=50, marker='x', label='Outliers')
plt.title("DBSCAN Clustering on Moon-shaped Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=1.0, random_state=42)

# Step 2: Evaluate Silhouette Scores for k = 2 to 5
sil_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)

# Step 3: Plot Silhouette Scores as a bar chart
plt.figure(figsize=(8,6))
plt.bar(k_values, sil_scores, color='skyblue')
plt.xticks(k_values)
plt.xlabel("Number of clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different K in K-Means")
plt.show()
```

Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Step 1: Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Create meshgrid for decision boundary visualization
h = 0.1  # step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 4: Plot decision boundaries and points
cmap_background = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_points = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, cmap=cmap_background, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap=cmap_points, s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='yellow', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Load the Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Compute inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Step 4: Plot inertia values
plt.figure(figsize=(8,6))
plt.plot(k_values, inertia_values, marker='o', linestyle='--', color='blue')
plt.xticks(k_values)
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")
plt.title("K-Means Inertia for Breast Cancer Dataset")
plt.show()
```

 Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding
noise)

```python
# Import required libraries
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Load the Wine dataset
wine = load_wine()
X = wine.data

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
y_db = dbscan.fit_predict(X_scaled)

# Step 4: Count the number of clusters (excluding noise)
n_clusters = len(set(y_db)) - (1 if -1 in y_db else 0)
print("Number of clusters (excluding noise):", n_clusters)

# Optional: Count number of noise points
n_noise = np.sum(y_db == -1)
print("Number of noise points:", n_noise)
```

 Generate synthetic data with make_blobs and apply KMeans. Then plot the cluster centers on top of the
data points

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Visualize clusters and cluster centers
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, label='Data points')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering with Cluster Centers")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

Load the Iris dataset, cluster with DBSCAN, and print how many samples were identified as noise

```python
# Import required libraries
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)
y_db = dbscan.fit_predict(X_scaled)

# Step 4: Count the number of noise points
n_noise = np.sum(y_db == -1)
print("Number of samples identified as noise:", n_noise)
```

 Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D
scatter plot

```python
# Import required libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data

# Step 2: Reduce dimensions to 3 using PCA
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X)

# Step 3: Apply K-Means clustering
kmeans = KMeans(n_clusters=10, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)

# Step 4: Visualize clusters in 3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y_kmeans, cmap='tab10', s=50)

# Plot centroids
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2],
           c='red', s=200, marker='X', label='Centroids')

ax.set_title("K-Means Clustering on Digits Dataset (3D PCA)")
ax.set_xlabel("PCA Component 1")
ax.set_ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
ax.legend()
plt.show()
```

Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the

```python
# Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic data with 5 centers
X, y_true = make_blobs(n_samples=500, centers=5, cluster_std=1.0, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Evaluate clustering using Silhouette Score
score = silhouette_score(X, y_kmeans)
print("Silhouette Score for K-Means clustering:", score)
```

Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering.
Visualize in 2D

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering

# Step 1: Load the Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Reduce dimensionality to 2D using PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Step 4: Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=2, linkage='ward')
y_agg = agg_clust.fit_predict(X_pca)

# Step 5: Visualize clusters in 2D
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_agg, cmap='viridis', s=50)
plt.title("Agglomerative Clustering on Breast Cancer Dataset (2D PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Step 1: Generate noisy circular data
X, y_true = make_circles(n_samples=300, noise=0.05, factor=0.5, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.15, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Step 4: Visualize results side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# K-Means plot
axes[0].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                c='red', s=200, marker='X', label='Centroids')
axes[0].set_title("K-Means Clustering on Circular Data")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")
axes[0].legend()

# DBSCAN plot
outliers = y_dbscan == -1
axes[1].scatter(X[y_dbscan != -1, 0], X[y_dbscan != -1, 1], c=y_dbscan[y_dbscan != -1], cmap='viridis', s=50)
axes[1].scatter(X[outliers, 0], X[outliers, 1], c='red', s=50, marker='x', label='Outliers')
axes[1].set_title("DBSCAN Clustering on Circular Data")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")
axes[1].legend()

plt.show()
```

Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
labels = kmeans.fit_predict(X)

# Step 3: Compute silhouette scores for each sample
sil_scores = silhouette_samples(X, labels)

# Step 4: Plot Silhouette Coefficient for each sample
plt.figure(figsize=(8,6))
y_ticks = np.arange(len(sil_scores))
plt.barh(y_ticks, sil_scores, color='skyblue')
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample Index")
plt.title("Silhouette Coefficient for Each Sample (Iris Dataset)")
plt.show()

# Optional: Print average silhouette score
avg_score = silhouette_score(X, labels)
print("Average Silhouette Score:", avg_score)
```

Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features)

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Step 1: Load the Wine dataset
wine = load_wine()
X = wine.data[:, :4]  # Use first 4 features
feature_names = wine.feature_names[:4]

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Step 3: Create a DataFrame for Seaborn
df = pd.DataFrame(X, columns=feature_names)
df['Cluster'] = y_kmeans.astype(str)  # Convert cluster labels to string for coloring

# Step 4: Visualize clusters using pairplot
sns.pairplot(df, hue='Cluster', palette='Set2', diag_kind='kde', corner=False)
plt.suptitle("K-Means Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

 Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters

```python
# Import required libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Step 1: Load the Digits dataset
digits = load_digits()
X = digits.data

# Step 2: Reduce dimensions to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

# Step 3: Apply Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=10, linkage='ward')
y_agg = agg_clust.fit_predict(X_tsne)

# Step 4: Visualize clusters
plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_agg, cmap='tab10', s=50)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
