Question 1: What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a formal method of making decisions or inferences about a population based on sample data.

  It is used to determine whether there is enough statistical evidence to support a particular claim (called a hypothesis) about a population parameter.

- Key Ideas:

- Null Hypothesis (H₀):

  A default assumption or claim (e.g., "there is no effect" or "the population mean = 50").

  It represents the status quo or no difference.

- Alternative Hypothesis (H₁ or Ha):

  Competing claim that we want to test (e.g., "the mean is not 50" or "there is an effect").

- Test Statistic:

  A value calculated from the sample data (e.g., Z, t, χ², F) to measure how far the data are from what H₀ predicts.

- Significance Level (α):

  A threshold probability (commonly 0.05) that determines how much evidence is required to reject H₀.

- p-value:

  Probability of observing data as extreme (or more extreme) as the sample result if H₀ were true.

  If p ≤ α, reject H₀; otherwise, fail to reject H₀.

- Decision Making:

  Reject H₀ → conclude there is enough evidence for H₁.

  Fail to reject H₀ → not enough evidence against H₀.

Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?

- Null Hypothesis (H₀):

  The null hypothesis is the default or starting assumption in hypothesis testing.

  It usually states that there is no effect, no difference, or no relationship in the population.

  It is the hypothesis we test against, to see if the data provide enough evidence to reject it.

- Alternative Hypothesis (H₁ or Ha):

  The alternative hypothesis is what we want to provide evidence for.

  It states that there is an effect, a difference, or a relationship in the population.

  It is accepted only if data provide strong enough evidence against H₀.

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test

- Significance Level (α)

  The significance level, denoted by α (alpha), is a threshold set by the researcher before conducting a hypothesis test.

  It represents the maximum probability of making a Type I error (rejecting a true null hypothesis).

  Common values: 0.05 (5%), 0.01 (1%), or 0.10 (10%).

Question 4: What are Type I and Type II errors? Give examples of each

- Type I Error (False Positive)

  Occurs when we reject the null hypothesis (H₀) when it is actually true.

  It means we detect an effect that doesn’t actually exist.

  Probability of committing a Type I error = α (significance level).

- Type II Error (False Negative)

  Occurs when we fail to reject the null hypothesis (H₀) when it is actually false.

  It means we miss a real effect.

  Probability of committing a Type II error = β.

  The power of a test = 1 – β (ability to correctly detect a true effect).

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.
-  Z-test

  Used when the population variance (σ²) or standard deviation (σ) is known.

  Works well with large sample sizes (n ≥ 30), due to the Central Limit Theorem.

  Based on the standard normal distribution (Z-distribution).

- T-test

  Used when the population variance (σ²) is unknown, and we rely on the sample variance (s²) as an estimate.

  Commonly used for small sample sizes (n < 30).

  Based on the Student’s t-distribution, which has heavier tails than the normal distribution (to account for extra uncertainty from estimating σ).



Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10     # number of trials
p = 0.5    # probability of success
size = 10000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', density=True, alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python
import numpy as np
from scipy.stats import norm

# Sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Step 1: Hypotheses
mu0 = 50      # population mean (null hypothesis)
sigma = 1     # assumed known population standard deviation
alpha = 0.05  # significance level

# Step 2: Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Step 3: Compute Z-statistic
z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))

# Step 4: Compute p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Step 5: Print results
print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Step 6: Decision
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)   # for reproducibility
mu, sigma = 50, 5    # true mean and standard deviation
n = 100              # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Step 3: Compute 95% confidence interval using t-distribution
alpha = 0.05
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)  # two-tailed
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print("Sample Mean:", sample_mean)
print("95% Confidence Interval: [{:.2f}, {:.2f}]".format(ci_lower, ci_upper))

# Step 5: Plot histogram
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores and plot histogram
def z_score_analysis(data):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std deviation

    # Step 2: Compute Z-scores
    z_scores = (data - mean) / std

    # Step 3: Print first few Z-scores
    print("First 10 Z-scores:", z_scores[:10])

    # Step 4: Plot histogram of standardized data
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, density=True)
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

    return z_scores

# Example dataset
data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5])

# Run function
z_scores = z_score_analysis(data)
```
