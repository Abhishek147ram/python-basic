Question 1: What are Large Language Models (LLMs) and how do they function?
- Large Language Models (LLMs) are advanced AI systems trained on massive amounts of text data to understand and generate human-like language. Built using deep learning techniques—especially the Transformer architecture—they learn patterns in language such as grammar, context, and meaning by predicting the next word in a sequence during training. Once trained, LLMs like ChatGPT can perform tasks such as answering questions, writing content, translating languages, and even coding, by analyzing input prompts and generating relevant, coherent responses based on the knowledge they have learned.

Question 5: Compare and contrast Langchain and LamaIndex. What unique problems
does each solve?
- LangChain and LlamaIndex are both popular frameworks used to build applications with large language models, but they solve different core problems and are often complementary rather than direct competitors. LangChain is designed as a **general-purpose application framework** that helps developers build end-to-end LLM-powered systems by chaining models, tools, APIs, memory, and agents into complex workflows—making it ideal for use cases like chatbots, autonomous agents, and multi-step reasoning systems. In contrast, LlamaIndex focuses specifically on the **data problem in LLMs**, enabling efficient data ingestion, indexing, and retrieval from large, diverse datasets, which is crucial for building Retrieval-Augmented Generation (RAG) systems such as document Q&A or knowledge bases. While both support RAG, LlamaIndex excels in **fast and accurate information retrieval from structured and unstructured data**, whereas LangChain excels in **orchestrating logic, workflows, and interactions between multiple components**. In summary, LlamaIndex solves the challenge of “how to connect and retrieve the right data for LLMs,” while LangChain solves “how to build intelligent applications and workflows using LLMs,” and in many real-world systems, they are used together—LlamaIndex for the data layer and LangChain for the application layer.



Question 6: Implement a basic Langchain pipeline using OpenAI’s LLM to answer
questions based on a user input prompt.
(Include your Python code and output in the code box below.)

```python
# Install required libraries (run once)
# pip install langchain openai

from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import os

# Set your OpenAI API Key
os.environ["OPENAI_API_KEY"] = "your_api_key_here"

# Step 1: Initialize the LLM
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

# Step 2: Create a Prompt Template
template = """
You are an intelligent assistant. Answer the following question clearly:

Question: {question}
Answer:
"""

prompt = PromptTemplate(
    input_variables=["question"],
    template=template
)

# Step 3: Create the LangChain pipeline
chain = LLMChain(llm=llm, prompt=prompt)

# Step 4: Take user input
user_question = input("Enter your question: ")

# Step 5: Run the pipeline
response = chain.run(user_question)

# Step 6: Print output
print("\nAnswer:\n", response)
```

Question 7: Integrate Langchain with a third-party API (e.g., weather, news) and show
how responses can be generated via LLMs.
(Include your Python code and output in the code box below.)

```python
id="langchain-weather-api"
# Install required libraries
# pip install langchain openai requests

import os
import requests
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# सेट करें अपनी API keys
os.environ["OPENAI_API_KEY"] = "your_openai_api_key"
WEATHER_API_KEY = "your_openweather_api_key"

# Step 1: Function to fetch weather data
def get_weather(city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={WEATHER_API_KEY}&units=metric"
    response = requests.get(url)
    data = response.json()

    if data.get("main"):
        temp = data["main"]["temp"]
        desc = data["weather"][0]["description"]
        return f"Temperature: {temp}°C, Condition: {desc}"
    else:
        return "City not found"

# Step 2: Initialize LLM
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

# Step 3: Prompt Template
template = """
You are a smart assistant. Based on the weather data below, answer the user's query.

Weather Data: {weather}
User Question: {question}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["weather", "question"],
    template=template
)

# Step 4: Create chain
chain = LLMChain(llm=llm, prompt=prompt)

# Step 5: User input
city = input("Enter city name: ")
question = input("Ask something about the weather: ")

# Step 6: Fetch weather data
weather_data = get_weather(city)

# Step 7: Run LLM
response = chain.run({
    "weather": weather_data,
    "question": question
})

# Step 8: Output
print("\nGenerated Answer:\n", response)
```

Question 9: Demonstrate combining Langchain with LamaIndex to create a simple
document-based Q&A chatbot.
(Include your Python code and output in the code box below.)

```python
id="langchain-llamaindex-chatbot"
# Install required libraries
# pip install llama-index langchain openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.core import ServiceContext
from llama_index.llms.openai import OpenAI as LlamaOpenAI

from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Step 1: Set API Key
os.environ["OPENAI_API_KEY"] = "your_openai_api_key"

# Step 2: Load local documents
documents = SimpleDirectoryReader("data").load_data()

# Step 3: Create LlamaIndex (for retrieval)
llama_llm = LlamaOpenAI(model="gpt-3.5-turbo", temperature=0)
service_context = ServiceContext.from_defaults(llm=llama_llm)

index = VectorStoreIndex.from_documents(documents, service_context=service_context)

# Step 4: Create query engine (retriever)
query_engine = index.as_query_engine()

# Step 5: LangChain LLM (for final response generation)
langchain_llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

template = """
You are a helpful assistant. Use the context below to answer the question.

Context:
{context}

Question:
{question}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=template
)

chain = LLMChain(llm=langchain_llm, prompt=prompt)

# Step 6: Chat loop
print("📄 Document Q&A Chatbot (type 'exit' to quit)\n")

while True:
    user_query = input("You: ")

    if user_query.lower() == "exit":
        break

    # Step 7: Retrieve relevant context using LlamaIndex
    retrieved_response = query_engine.query(user_query)
    context = str(retrieved_response)

    # Step 8: Generate final answer using LangChain
    final_answer = chain.run({
        "context": context,
        "question": user_query
    })

    print("\nBot:", final_answer, "\n")
```

Question 10: A legal firm wants to use AI to summarize large volumes of legal
documents and retrieve relevant information quickly. Propose a solution using
Langchain and LamaIndex, and explain how it would work in practice.
(Include your Python code and output in the code box below.)
- A practical solution for a legal firm is to build a Retrieval-Augmented Generation (RAG) system that uses LlamaIndex for document ingestion, indexing, and retrieval and LangChain for orchestration, prompting, and summarization workflows. In practice, large volumes of legal documents (contracts, case files, judgments) are first loaded and chunked into smaller sections. LlamaIndex converts these into embeddings and stores them in a vector index for fast semantic search. When a lawyer asks a query (e.g., “Summarize key clauses in this contract” or “Find precedents related to breach of agreement”), LlamaIndex retrieves the most relevant chunks. LangChain then takes this retrieved context and applies structured prompts (for summarization, Q&A, or extraction of clauses) using an LLM to generate clear, concise answers. This approach ensures speed (via indexing), accuracy (via context retrieval), and usability (via natural language responses), making it ideal for legal research and document analysis.

```python
id="legal-rag-system"
# Install required libraries
# pip install llama-index langchain openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.openai import OpenAI as LlamaOpenAI
from llama_index.core import ServiceContext

from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Step 1: Set API Key
os.environ["OPENAI_API_KEY"] = "your_openai_api_key"

# Step 2: Load legal documents from local folder
documents = SimpleDirectoryReader("legal_docs").load_data()

# Step 3: Create LlamaIndex (for indexing & retrieval)
llama_llm = LlamaOpenAI(model="gpt-3.5-turbo", temperature=0)
service_context = ServiceContext.from_defaults(llm=llama_llm)

index = VectorStoreIndex.from_documents(documents, service_context=service_context)

# Step 4: Create query engine
query_engine = index.as_query_engine()

# Step 5: LangChain LLM for summarization & response generation
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.5)

template = """
You are a legal AI assistant. Use the provided context to:
1. Summarize the document OR
2. Answer the legal query precisely

Context:
{context}

User Query:
{query}

Provide a clear, professional response:
"""

prompt = PromptTemplate(
    input_variables=["context", "query"],
    template=template
)

chain = LLMChain(llm=llm, prompt=prompt)

# Step 6: User query loop
print("⚖️ Legal Document AI Assistant (type 'exit' to quit)\n")

while True:
    user_query = input("Enter your query: ")

    if user_query.lower() == "exit":
        break

    # Step 7: Retrieve relevant legal context
    retrieved = query_engine.query(user_query)
    context = str(retrieved)

    # Step 8: Generate final answer
    response = chain.run({
        "context": context,
        "query": user_query
    })

    print("\nAI Response:\n", response, "\n")
```
