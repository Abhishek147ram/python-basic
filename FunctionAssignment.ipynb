7. What are the advantages of using generators over regular functions?
 - Generators yield one item at a time, so they don’t store the entire result in memory.
 - Values are produced only when needed, which speeds up performance in some cases.
 - Automatically remembers where it left off between yield calls.
 - Easier to write than manually creating iterator classes.

 - map(): Applies a function to each item in an iterable.

 Example: filter(lambda x: x%2==0, [1, 2, 3]) - [2]

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def greet(n):
  return sum(num for num in n if num%2==0)
greet([11,12,13,14,15,16,17,18,19])
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def greet(n):
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def greet(n):
  for i in range(n+1):
    yield 2**i
```

```python
greet(2)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def greet(n):
    with open(n, 'r') as l:
        for line in l:
            yield line.rstrip('\n')
```

```python
greet(12)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
c = [39.2, 36.5, 37.3, 38, 37.8]
f = list(map(lambda x: (x * 9/5) + 32, c))
print(f)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def greet(n):
  vowels = 'aeiouAEIOU'
  return filter(lambda x: x not in vowels, n)
```

```python
greet("Abhishek")
```

product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.


```python
orders = [(34587,4,40.95),(98762,5,56.80),(77336,3,32.95),(88112,3,24.99)]

result = list(map(lambda order: (order[0],(order[1] * order[2]) if (order[1] * order[2]) >= 100 else (order[1] * order[2] + 10) ), orders))

print(result)
```
