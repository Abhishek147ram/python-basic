Question 4: Describe the architecture and functioning of a Recurrent Natural Network
(RNN).
- A Recurrent Neural Network (RNN) is a type of neural network designed to process sequential data such as text, speech, or time-series.

- Architecture:
RNN has input, hidden, and output layers. The hidden layer has recurrent connections, meaning the output of the previous step is fed back into the network along with the current input.

- Functioning:
At each time step, the network takes the current input and the previous hidden state to produce a new hidden state and output. This allows the model to remember past information and understand context in sequences.

Question 5: What is the key difference between LSTM and GRU networks in NLP
applications?
- The key difference between LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) is in their architecture and complexity.

- LSTM: Uses three gates (input, forget, output) and a separate cell state to store long-term information. It is more complex but very powerful for capturing long dependencies.

- GRU: Uses two gates (update and reset) and combines the cell state with the hidden state. It is simpler and faster to train.

Question 7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download tokenizer (run once)
nltk.download('punkt')
nltk.download('punkt_tab') # Added to resolve LookupError for punkt_tab

# Sample paragraph
text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

# Tokenization
tokens = word_tokenize(text)

# Frequency Distribution
freq_dist = FreqDist(tokens)

print("Tokens:")
print(tokens)

print("\nTop 10 Most Common Words:")
print(freq_dist.most_common(10))
```

Question 8: Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0). # Dataset texts = [ “I love this project”, #Positive “This is an amazing experience”, #Positive “I hate waiting in line”, #Negative “This is the worst service”, #Negative “Absolutely fantastic!” #Positive ] labels = [1, 1, 0, 0, 1] Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.

```python
# Import required libraries
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
"I love this project",              # Positive
"This is an amazing experience",    # Positive
"I hate waiting in line",           # Negative
"This is the worst service",        # Negative
"Absolutely fantastic!"             # Positive
]

labels = [1, 1, 0, 0, 1]

# Tokenization
tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding sequences
max_length = 5
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# Convert labels to numpy array
labels = np.array(labels)

# Build LSTM Model
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=8, input_length=max_length))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(padded_sequences, labels, epochs=10, verbose=1)

# Evaluate the model
loss, accuracy = model.evaluate(padded_sequences, labels)
print("Accuracy:", accuracy)
```

Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:
“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.”
Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.


```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Paragraph
text = """Homi Jehangir Bhabha was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

# Process text
doc = nlp(text)

# Print Tokens and Lemmas
print("Tokens and Lemmas:")
for token in doc:
    print(token.text, "->", token.lemma_)

# Print Named Entities
print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

Question 10: You are working on a chatbot for a mental health platform. Explain how
you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
NLP to understand and respond to user input effectively. Detail your architecture, data
preprocessing pipeline, and any ethical considerations.

-To build a mental health chatbot, we can combine text processing libraries with deep learning models.

1. Architecture

 Input Layer: User message.

Preprocessing Layer: Use spaCy for tokenization, lemmatization, and entity recognition.

Sequence Model: Use LSTM or GRU to understand context and sequential patterns in sentences.

Dense Layer + Softmax: Classifies the intent (e.g., stress, anxiety, greeting).

Response Generator: Maps predicted intent to an appropriate supportive response.

2. Data Preprocessing Pipeline

Collect conversation dataset (mental health queries).

Clean text (remove punctuation, lowercase).

Use spaCy for tokenization and lemmatization.

Convert words to sequences using Tokenizer.

Apply padding to make equal-length sequences.

Train an LSTM/GRU model for intent classification.

3. Ethical Considerations

Ensure privacy and data protection of user conversations.

Avoid giving medical diagnoses; only provide supportive guidance.

Include a human support escalation option.

Ensure responses are empathetic, unbiased, and safe.

```python
import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Example dataset
texts = [
"I feel very stressed and tired",
"I am happy today",
"I feel anxious about my exams",
"Hello, how are you?",
"I feel very sad and lonely"
]

labels = [1, 0, 1, 2, 1]
# 0 = positive mood, 1 = distress, 2 = greeting

# Preprocessing using spaCy (lemmatization)
processed_texts = []
for text in texts:
    doc = nlp(text)
    processed_texts.append(" ".join([token.lemma_ for token in doc]))

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)
sequences = tokenizer.texts_to_sequences(processed_texts)

# Padding
X = pad_sequences(sequences, maxlen=6)

y = np.array(labels)

# LSTM Model
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=16, input_length=6))
model.add(LSTM(32))
model.add(Dense(3, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',
optimizer='adam',
metrics=['accuracy'])

# Train model
model.fit(X, y, epochs=10, verbose=1)

# Test input
test_text = "I feel very nervous today"
doc = nlp(test_text)
processed = " ".join([token.lemma_ for token in doc])

seq = tokenizer.texts_to_sequences([processed])
pad = pad_sequences(seq, maxlen=6)

prediction = model.predict(pad)
print("Predicted Class:", np.argmax(prediction))
```
